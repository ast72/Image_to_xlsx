```python
import os
import pandas as pd
import numpy as np
from PIL import Image
import xlsxwriter
```

```python
fichier = 'clown_fish.png'
img = Image.open(fichier)
img_data = np.array(img)
nom =  os.path.basename(fichier).split('.')[0]
print(nom)
img
```

```python
workbook = xlsxwriter.Workbook(fichier + '.xlsx')
worksheet = workbook.add_worksheet(nom)

for ligne in range(0,img_data.shape[0]):
    for colonne in range(0,img_data.shape[1]):
        for couleur in range(0,3):
            worksheet.write((ligne * 3) + couleur, colonne, img_data[ligne, colonne, couleur])
    
    worksheet.conditional_format((ligne * 3),0,ligne,img_data.shape[1],{'type': '2_color_scale', 
                                                                   'min_value': 0,
                                                                   'max_value': 255, 
                                                                   'min_color': "black", 
                                                                   'max_color': "red"})
    worksheet.conditional_format((ligne * 3) + 1,0,ligne + 1,img_data.shape[1],{'type': '2_color_scale', 
                                                                   'min_value': 0,
                                                                   'max_value': 255, 
                                                                   'min_color': "black", 
                                                                   'max_color': "green"})
    worksheet.conditional_format((ligne * 3) + 2,0,ligne + 2,img_data.shape[1],{'type': '2_color_scale', 
                                                                   'min_value': 0,
                                                                   'max_value': 255, 
                                                                   'min_color': "black", 
                                                                   'max_color': "blue"})
            
workbook.close()
```
